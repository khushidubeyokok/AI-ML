#Decision Trees

A Decision Tree is a supervised learning model used for classification or regression, where data is split at each node based on feature values to reduce uncertainty and reach a prediction

At every step, the model chooses the best feature to split on by measuring impurity using metrics like Entropy (from Information Gain), Gini Impurity, or Mean Squared Error (for regression).

It continues splitting until the data is pure or some stopping condition is met (like max depth or min samples).

They're fast and interpretable but unstable when used alone, small changes in data can change the structure

In this notebook we will classify salary of employees based on company,job and degree

```python
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv('/content/salaries.csv')
```

```python
dataset.head()
```

```python
x=dataset.drop('salary_more_then_100k',axis='columns')
y=dataset['salary_more_then_100k']
```

```python
#encoding text data into numbers
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
x['company_n'] = le_company.fit_transform(x['company'])
x['job_n'] = le_job.fit_transform(x['job'])
x['degree_n'] = le_degree.fit_transform(x['degree'])
```

```python
x.head()
```

```python
X=x.drop(['company','job','degree'],axis='columns')
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X,y) #default criterion='gini'
```

```python
model.score(X,y) # 1 because our dataset is simple, and score is for training data itself
```

```python
model.predict([[2,2,1]])
```
